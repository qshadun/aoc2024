```python
from collections import deque, defaultdict

def parse_input(input_file):
    grid = []
    with open(input_file) as f:
        for line in f:
            grid.append([c for c in line.rstrip()])
    return grid

def find_start(grid):
    m, n = len(grid), len(grid[0])
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 'S':
                return i, j

def find_end(grid):
    m, n = len(grid), len(grid[0])
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 'E':
                return i, j

def copy_grid(grid):
    ans = []
    for row in grid:
        ans.append(row[:])
    return ans

def print_grid(grid):
    for row in grid:
        print(''.join(row))

deltas = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def shorted_path(grid, start, bound):
    m, n = len(grid), len(grid[0])
    q = deque([start])
    visited = set(start)
    steps = 0
    while q:
        cur_len = len(q)
        for _ in range(cur_len):
            x, y = q.popleft()
            for dx, dy in deltas:
                nx, ny = x + dx, y + dy
                try:
                    if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] != '#' and (nx, ny) not in visited:
                        if grid[nx][ny] == 'E':
                            return steps + 1
                        q.append((nx, ny))
                        visited.add((nx, ny))
                except:
                    
                    print(nx, ny)
                    print(m, n)
                    print(len(grid[nx]))
        steps += 1
        if steps >= bound:
            return bound + 1000
            

def part1(input_file, cut_steps):
    grid = parse_input(input_file)
    start = find_start(grid)
    
    original = shorted_path(grid, start, 100000)
    print(original)
    m, n = len(grid), len(grid[0])
    ans = 0
    for i in range(1, m-1):
        for j in range(1, n-1):
            if grid[i][j] == '#':
                new_grid = copy_grid(grid)
                new_grid[i][j] = '.'
                if shorted_path(new_grid, start, original-cut_steps) <= original - cut_steps:
                    ans += 1

    return ans


def find_shorted_path(grid, start):
    m, n = len(grid), len(grid[0])
    q = deque([start])
    pre = {start: None}
    steps = 0
    while q:
        cur_len = len(q)
        for _ in range(cur_len):
            x, y = q.popleft()
            for dx, dy in deltas:
                nx, ny = x + dx, y + dy
                
                if 0 <= nx < m and 0 <= ny < n and grid[nx][ny] != '#' and (nx, ny) not in pre:
                    
                    q.append((nx, ny))
                    pre[(nx, ny)] = (x, y)
                    if grid[nx][ny] == 'E':
                        return steps + 1, pre
                
        steps += 1

def to_dist(pre, end):
    dist = {}
    cur = end
    cur_dist = 0
    while cur:
        dist[cur] = cur_dist
        cur_dist += 1
        cur = pre[cur]
    return dist

def part2(input_file, cut):
    grid = parse_input(input_file)
    m, n = len(grid), len(grid[0])
    start = find_start(grid)
    end = find_end(grid)
    steps, pre = find_shorted_path(grid, start)
    dist = to_dist(pre, end)
    ans = defaultdict(int)
    # def bfs(p1):
    #     visited = {p1}
    #     steps = 1
    #     q = deque([p1])
    #     d = dist[p1]
    #     while q and steps <= 20:
    #         cur_len = len(q)
    #         for _ in range(cur_len):
    #             x, y = q.popleft()
    #             for dx, dy in deltas:
    #                 nx, ny = x + dx, y + dy
    #                 if 1 <= nx < m-1 and 1 <= ny < n-1 and (nx, ny) not in visited:
                        
    #                     visited.add((nx, ny))
                        
    #                     if dist.get((nx, ny), d) <= d - cut - steps and steps >=2:
    #                         ans[d - dist.get((nx, ny)) - steps] += 1
    #                     if (nx, ny) != end:
    #                         q.append((nx, ny))
                        
    #         steps += 1
            

    for (x, y), d in dist.items():
        if d > cut:
            for dx in range(-20, 21):
                nx = x + dx
                if 1 <= nx < m - 1:
                    y_bound = 20 - abs(dx)
                    for dy in range(-y_bound, y_bound+1):
                        ny = y + dy
                        if 1 <= ny < n-1:
                            steps = abs(dx) + abs(dy)
                            if dist.get((nx, ny), d) <= d - cut - steps and steps >=2:
                                ans[d - dist.get((nx, ny)) - steps] += 1
            
    
    return ans, sum(ans.values())
```
