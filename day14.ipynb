```python
import re

def parse_input(input_file):
    ans = []
    with open(input_file) as f:
        for line in f:
            nums = re.findall(r'-?\d+', line.rstrip())
            ans.append([int(x) for x in nums])
    return ans

def move_one_direction(px, vx, bx):
    px += vx
    if px < 0:
        px += bx
    elif px >= bx:
        px -= bx
    return px

def move(robot, bx, by):
    px, py, vx, vy = robot
    return [move_one_direction(px, vx, bx), move_one_direction(py, vy, by), vx, vy]

def safety_score(robots, bx, by):
    s1, s2, s3, s4 = 0, 0, 0, 0
    mx = bx // 2
    my = by // 2
    for px, py, _, _ in robots:
        
        if px == mx or py == my:
            continue
        if 0 <= px < mx and 0 <= py < my:
            s1 += 1
        elif 0 <= px < mx and my < py:
            s2 += 1
        elif py < my:
            s3 += 1
        else:
            s4 += 1
    print(s1, s2, s3, s4)
    return s1 * s2 * s3 * s4

def part1(input_file, bx, by):
    robots = parse_input(input_file)
    # print(robots)
    end_robots = []
    for robot in robots:
        for _ in range(100):
            robot = move(robot, bx, by)
        end_robots.append(robot)
        
    return safety_score(end_robots, bx, by)

def to_grid(robots, bx, by):
    grid = [['.'] * bx for _ in range(by)]
    for px, py, _, _ in robots:
        grid[py][px] = '+'
    return grid

def print_grid(grid):
    for row in grid:
        print(''.join(row))

def check_tree(grid):
    m, n = len(grid), len(grid[0])
    staight_line_each_row = [0] * m
    for i in range(m):
        row = grid[i]
        left, right = 0, 0
        while left < n and right < n:
            while left < n and row[left] != '+':
                left += 1
            right = left + 1
            while right < n and row[right] == '+':
                right += 1
            staight_line_each_row[i] = max(staight_line_each_row[i], right - left)
            left = right
    return max(staight_line_each_row) > 10

def part2(input_file, bx, by):
    robots = parse_input(input_file)
    ans = 0
    while True:
        ans += 1
        for i in range(len(robots)):
            robots[i] = move(robots[i], bx, by)
        grid = to_grid(robots, bx, by)
        if check_tree(grid):
            print_grid(grid)
            return ans
```
